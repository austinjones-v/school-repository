```python
#9.1

#INPUTS
N = int(input('Enter N :  '))
dt = float(input('Enter dt:  '))
k = float(input('Enter k :  '))
m = float(input('Enter m :  '))
x = float(input('Enter x0:  '))
v = float(input('Enter v0:  '))

# START OFF TIME, POSITION, VELOCITY ARRAYS
time =array('f', [0.])
xsave=array('f',[x])
vsave=array('f',[v])

# DO THE MD AND APPEND EACH NEW TIME, POSITION, VELOCITY TO ITS ARRAY
for i in range (1,N):
    t=dt*i
    x=x+v*dt
    v=v-k*x*dt/m
    time.append(t)
    xsave.append(x)
    vsave.append(v)

# PRINT OUT TIME AND POSITION
for i in range (1,N):
    print(time[i],xsave[i])

# MAKE A PLOT
plt.figure(1)
plt.subplot(211)
plt.plot(time, xsave)
plt.xlabel('t')
plt.ylabel('x')
plt.subplot(212)
plt.plot(time, vsave)
plt.xlabel('t')
plt.ylabel('v')
plt.show()
```

```python
#9.4

#INPUTS
N = int(input('Enter N :  '))
dt = float(input('Enter dt:  '))
k = float(input('Enter k :  '))
m = float(input('Enter m :  '))
x1 = float(input('Enter x10:  '))
x2 = float(input('Enter x20:  '))
x3 = float(input('Enter x30:  '))
v1 = float(input('Enter v0:  '))
v2=v3=v1

# START OFF TIME, POSITION, VELOCITY ARRAYS
time =array('f', [0.])
x1save=array('f',[x1])
x2save=array('f',[x2])
x3save=array('f',[x3])
v1save=array('f',[v1])
v2save=array('f',[v2])
v3save=array('f',[v3])

# DO THE MD AND APPEND EACH NEW TIME, POSITION, VELOCITY TO ITS ARRAY
for i in range (1,N):
    t=dt*i
    x1=x1+v1*dt
    x2=x2+v2*dt
    x3=x3+v3*dt
    v1=v1-k*x1*dt/m
    v2=v2-k*x2*dt/m
    v3=v3-k*x3*dt/m
    time.append(t)
    x1save.append(x1)
    x2save.append(x2)
    x3save.append(x3)
    


# PRINT OUT TIME AND POSITION
for i in range (1,N):
    print(time[i],x1save[i])
    print(time[i],x2save[i])
    print(time[i],x3save[i])


# MAKE A PLOT
plt.plot(time,x1save,'k-', label = 'x1')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.plot(time,x2save,'b-',label = 'x2')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.plot(time,x3save,'r-',label = 'x3')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```
