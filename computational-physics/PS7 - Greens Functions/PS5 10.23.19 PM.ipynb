```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path1 = '/Users/Austin/Desktop/PHY 40/Homework/5/slughorn5_2.txt'

x, y = np.loadtxt(file_path1, unpack = True) #Method to extract/import information from txt file  

# print(min(x), max(x))
# print(min(y), max(y))
    
        
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('density')
#plt.legend()
ax=plt.gca() #Allows for axis limits
ax.set(xlim=(min(x), max(x)), ylim=(min(y), 1.05*max(y)))
plt.grid()
plt.show()
```
